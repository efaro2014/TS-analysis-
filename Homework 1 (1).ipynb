```python
#Q6
from random import gauss
from random import seed
from pandas import Series

# An simulation example of random walk
from random import random
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.3 else 1)
for i in range(1, 1000):
 movement = -1 if random() < 0.3 else 1
 value = random_walk[i-1] + movement
 random_walk.append(value)
plt.plot(random_walk)
plt.show()
```

```python
def plotMovingAverage(series, window, plot_actual=False, scale=1.96):

    rolling_mean = series.rolling(window=window).mean()

    plt.figure(figsize=(15,5))
    plt.title("Moving average\n window size = {}".format(window))
    plt.plot(rolling_mean, "g", label="Rolling mean trend")
    
    if plot_actual:
        plt.plot(series[window:], label="Actual values")
        
    plt.legend(loc="upper left") 
    plt.grid(True)
```

```python
t=np.arange(0,1000)
df=pd.DataFrame()
df['walk']=np.array(random_walk)
df['t']=t
plotMovingAverage(df.walk, 30)  
plotMovingAverage(df.walk, 365)  
```
